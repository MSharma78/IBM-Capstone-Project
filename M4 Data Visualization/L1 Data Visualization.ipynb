**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
QUERY = """
SELECT CompTotal
FROM main
"""

df = pd.read_sql_query(QUERY, conn)

df = df[df['CompTotal'] < 1000000]

plt.hist(df['CompTotal'], bins=30)
plt.title("Distribution of Total Compensation")
plt.xlabel("CompTotal")
plt.ylabel("Frequency")
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = """
SELECT AGE FROM MAIN
"""
df = pd.read_sql(QUERY, conn)

age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df['Age'] = df['Age'].map(age_map)

plt.boxplot(df['Age'].dropna())
plt.title('Age Distribution')
plt.ylabel('Age')
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = """ SELECT AGE, WORKEXP from main """
df = pd.read_sql(QUERY, conn)

plt.scatter(df['WorkExp'], df['Age'])
plt.ylabel("Age")
plt.xlabel('WorkExperience')
plt.title('Age vs WorkExp')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
QUERY = """SELECT TimeSearching, Frustration, Age FROM main"""
df = pd.read_sql(QUERY, conn)

df['Frustration_count'] = df['Frustration'].apply(lambda x: len(str(x).split(';')))
df = df[df['Frustration_count'] <= 5]

# Convert Age to numeric
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df['Age'] = df['Age'].map(age_map)

# Convert TimeSearching to numeric
time_map = {
    'Less than 15 minutes': 1,
    '15-30 minutes': 2,
    '30-60 minutes': 3,
    '60-120 minutes': 4,
    'More than 120 minutes': 5
}
df['TimeSearching'] = df['TimeSearching'].map(time_map)

# Convert Frustration to number of frustrations selected
df['Frustration_count'] = df['Frustration'].apply(lambda x: len(str(x).split(';')))

# Bubble plot
plt.scatter(
    df['TimeSearching'],
    df['Frustration_count'],
    s=df['Age']*3,
    alpha=0.5
)

plt.xlabel("Time Searching")
plt.ylabel("Number of Frustrations")
plt.title("Bubble Plot: Time Searching vs Frustration")

plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY = """SELECT DatabaseWantToWorkWith from main"""

df = pd.read_sql(QUERY, conn)

db_series = df['DatabaseWantToWorkWith'].str.split(';').explode()

# count occurrences
top5 = db_series.value_counts().head(5)


top5.plot(kind='pie', autopct='%1.1f%%')
plt.title("DATABASE Distribution")
plt.show()
```

```python
QUERY = """SELECT Age, TimeSearching, TimeAnswering FROM main"""
df = pd.read_sql(QUERY, conn)

# filter correct age group
df_filtered = df[df['Age'] == '25-34 years old']

# convert categories to numeric
time_map = {
    'Less than 15 minutes': 1,
    '15-30 minutes': 2,
    '30-60 minutes': 3,
    '60-120 minutes': 4,
    'More than 120 minutes': 5
}

df_filtered['TimeSearching'] = df_filtered['TimeSearching'].map(time_map)
df_filtered['TimeAnswering'] = df_filtered['TimeAnswering'].map(time_map)

# drop rows where mapping failed
df_filtered = df_filtered.dropna(subset=['TimeSearching','TimeAnswering'])

median_vals = df_filtered[['TimeSearching','TimeAnswering']].median()

median_vals.plot(kind='bar', stacked=True)

plt.title("Median Time Searching vs Answering (Age 25-34)")
plt.ylabel("Median Category Value")
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = """SELECT Age, CompTotal FROM main"""
df = pd.read_sql(QUERY, conn)

df['Age_num'] = df['Age'].map(age_map)

df_filtered = df[(df['Age_num'] >= 45) & (df['Age_num'] <= 60)]

median_comp = df_filtered.groupby('Age_num')['CompTotal'].median()

median_comp.plot(kind='line', marker='o')

plt.title("Median Compensation for Ages 45–60")
plt.xlabel("Age")
plt.ylabel("Median Compensation")
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
QUERY = """SELECT MainBranch FROM main"""
df = pd.read_sql(QUERY, conn)

df['MainBranch'].value_counts().plot(kind='barh')

plt.title("Distribution of Respondents by Main Branch")
plt.xlabel("Count")
plt.ylabel("Main Branch")

plt.show()
```
